###Dimensionality Reduction Techniques

Ideally, because there are so many variables in the dataset, it would be best to reduce the number of dimensions, to visualize the dataset, prevent overfitting and increase the speed of the model.

There are two ways this can be accomplished.

> Principal component Analysis (PCA)

> Linear Discriminant Analysis (LDA)








```python
# Normalizing numerical features so that each feature has mean 0 and variance 1
feature_scaler = StandardScaler()
X_scaled = feature_scaler.fit_transform(X)
```

```python
# Implementing PCA to visualize dataset
pca = PCA(n_components = 50) 
# PCA can make as many principle components as there are dimensions in the dataset
# That is exactly what would happen if the n_components was not specified
pca.fit(X_scaled)
x_pca = pca.transform(X_scaled)

print("Variance explained by each of the n_components: ",pca.explained_variance_ratio_)
print("Total variance explained by the n_components: ",sum(pca.explained_variance_ratio_))
```

```python
# Plotting a scatter graph to see how the first two Principal Components classify the data
digits=list(df['label'])
data = [go.Scatter(x=x_pca[:,0], y=x_pca[:,1], mode='markers',
                    marker = dict(color=Y, colorscale='rainbow', opacity=0.5),
                                text=[f'digit: {a}' for a in digits],
                                hoverinfo='text')]

layout = go.Layout(title = 'PCA Dimensionality Reduction', width = 700, height = 700,
                    xaxis = dict(title='First Principal Component'),
                    yaxis = dict(title='Second Principal Component'))
fig = go.Figure(data=data, layout=layout)
fig.show()
```

PCA is an unsupervised machine learning algorithm that does not take the target variable into acccount. It only finds the direction in data in which the data varies the most.

As seen in the output, the 50 principal components have captured 88% of the variance in the dataset.

However, when we plot the first two principal components against each other, we can see that because the target labels were not taken into account, they are not doing a good job of actually separating the classes

```python
# Implementing LDA to visualize dataset
lda = LDA()
# LDA can create as many linear discriminants as there are classes - 1
# max linear discriminants = n_classes - 1
# This is exactly what would take place n_components was not passed into the function
x_lda = lda.fit_transform(X_scaled, Y)

print("Variance explained by each of the n_components: ",lda.explained_variance_ratio_)
print("Total variance explained by the n_components: ",sum(lda.explained_variance_ratio_))
```

```python
# Plotting a scatter graph to see how the first two Linear Discriminants classify the data
data = [go.Scatter(x=x_lda[:,0], y=x_lda[:,1], mode='markers',
                    marker = dict(color=Y, colorscale='Rainbow', opacity=0.5),
                                text=[f'digit: {a}' for a in digits],
                                hoverinfo='text')]

layout = go.Layout(title = 'LDA Dimensionality Reduction', width = 700, height = 700,
                    xaxis = dict(title='Discriminant 1'),
                    yaxis = dict(title='Discriminant 2'))
fig = go.Figure(data=data, layout=layout)
fig.show()
```

LDA however is a supervised machine algorithm used in dimensionality reduction. It creates a new axis the maximises the distance between the means of each class and mininises the variance within each class.

The first output shows that the 23 linear discriminants account for 100% of the variance in the data.

Moreover, when the first two linear discriminants are plotted against each other the graph shows that the data is being very well classified. 

###Classification models

In this section three (3) classifiers will be implemented to model the data

> Support Vector Classifier

> Random Forest Classifier

> Artificial Neural Network: Multi-layer Perceptron (MLP) 

```python
from sklearn.model_selection import GridSearchCV
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

Implementing Support Vector Classifier

```python
# We are building a pipeline, which takes in SMOTE (Synthetic Minority Oversampling TEchnique) 
# and the Classification model, SVC (Support Vector Classifier).

# SVC a line to separate the data supported using only a few data points (vectors).
# SMOTE is being implented here because the dataset is imbalanced.

model = Pipeline([
        ('balancing', SMOTE(random_state = 101)),
        ('classification', SVC(random_state=1))
    ])

# Because we do not know the best parameters to use in developing the model, 
# we would run a grid search with a list of parameters and then pick the model
# that gives us the best score
# We are trying all the 4 kernels,
# and, classification__C here refers to regularization term (penalty factor) in SV machines

grid_param = {'classification__kernel': ['linear','poly','rbf','sigmoid'], 'classification__C': [.001,.01,.1,1,10,100]}

# Scoring was set to accuracy because we would want a model that will accurately predict
# the hand gestures correctly

# cv = 5 means we a re using 5 - fold cross validation

gd_sr = GridSearchCV(estimator=model, param_grid=grid_param, scoring='accuracy', cv=5)

# Fitting the model to our data

gd_sr.fit(x_lda, Y)

best_parameters = gd_sr.best_params_ # This shows us the parameters that gives us the best score
print("Best parameters: ", best_parameters)

best_result = gd_sr.best_score_ # Mean cross-validated score of the best_estimator
print("Best result: ", best_result)
```

Implementing Decision Tree Classifier

```python
# We are building a pipeline, which takes in SMOTE (Synthetic Minority Oversampling TEchnique) 
# and the Classification model, Random Forest Classifier.
# SMOTE is being implented here because the dataset is imbalanced.

# We are using entropy as the crieteria for constructing out tree
# Meaning that the split that gives us the best information gain
# will be used for the split

# max_features has been set to auto, this means thst
# The square root of the total number if features will be used considered when
# looking for the bestsplit

model = Pipeline([
        ('balancing', SMOTE(random_state = 101)),
        ('classification', RandomForestClassifier(criterion='entropy', max_features='auto', random_state=1) )
    ])

# n_estimators here represents the total number of trees that a developed in parallel.
# As a general rule of thumb this number should not exceed 100.
# That being said, although we are using [75,100,125] as our parameter,
# we had run the code before using [10,20,30,40,50,100] and 100 was selected as the best parameter.

grid_param = {'classification__n_estimators': [75,100,125]}

gd_sr = GridSearchCV(estimator=model, param_grid=grid_param, scoring='accuracy', cv=5)

gd_sr.fit(x_lda, Y)

best_parameters = gd_sr.best_params_ # This shows us the parameters that gives us the best score
print(best_parameters)

best_result = gd_sr.best_score_ # Mean cross-validated score of the best_estimator
print(best_result)
```

Implementing Multi-layer Perceptron (MLP)

```python
# We are building a pipeline, which takes in SMOTE (Synthetic Minority Oversampling TEchnique) 
# and the Classification model, Multi-layer Perceptron Classifier.
# SMOTE is being implented here because the dataset is imbalanced.

model = Pipeline([
        ('balancing', SMOTE(random_state = 101)),
        ('classification', MLPClassifier(random_state=1)) 
    ])

# We are using four hyper parameters
# classification__activation: the activation functions of the hidden layer(s)
# classification__learning_rate_init: controls the step-size in updating the weights.
# classification__max_iter: the maximum number of iterations
# classification__hidden_layer_sizes: the number of hidden layers and nodes in each layer
# (100,) - one hidden layer with 100 nodes
# (100,100) - two hidden layers with 100 nodes each

grid_param = {'classification__activation': ['identity','logistic', 'tanh', 'relu'],
              'classification__learning_rate_init': [0.0001, 0.001, 0.01, 0.1, 1, 10], 'classification__max_iter' : [750,1000,1250],
              'classification__hidden_layer_sizes': [(100,),(100,100)]}

gd_sr = GridSearchCV(estimator=model, param_grid=grid_param, scoring='accuracy', cv=5)

gd_sr.fit(x_lda, Y)

best_parameters = gd_sr.best_params_ # This shows us the parameters that gives us the best score
print(best_parameters)

best_result = gd_sr.best_score_ # Mean cross-validated score of the best_estimator
print(best_result)
```

All three of the models are performing excellently, doing a good job of accurately predicting the hand gestures.

The best performing model is the Multi-layer perceptron with an accuracy score of 99.92%

The reason all models are giving good scores could be that the Linear Discriminant accounts for all the variance in the data. Meaning, all the data points have a very clear separation.
